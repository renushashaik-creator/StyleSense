```python
# Install libraries
!pip install diffusers transformers accelerate gradio pillow

import torch
from diffusers import StableDiffusionInpaintPipeline
from PIL import Image
import gradio as gr

# Load Stable Diffusion model
pipe = StableDiffusionInpaintPipeline.from_pretrained(
    "runwayml/stable-diffusion-inpainting",
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")

def virtual_tryon(model_img, garment_img):
    prompt = "A person wearing the given clothing naturally, realistic, high quality"

    model_img = model_img.resize((512, 512))
    garment_img = garment_img.resize((512, 512))

    result = pipe(
        prompt=prompt,
        image=model_img,
        mask_image=garment_img
    ).images[0]

    return result

interface = gr.Interface(
    fn=virtual_tryon,
    inputs=[
        gr.Image(type="pil", label="Upload Model Image"),
        gr.Image(type="pil", label="Upload Garment Image")
    ],
    outputs=gr.Image(label="Generated Try-On Image"),
    title="AI Virtual Try-On System"
)

interface.launch()
```
